```python
#random data shuffelin
data.sample(frac=1)
#Spliting the data
X = data.drop("Outcome",axis=1)
y = data["Outcome"]
X_train,X_test,y_train,y_test =  train_test_split(X,y,test_size=0.2)
```

```python
## Train the model (Logistic Regression)
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression(random_state=0)
log_reg.fit(X_train,y_train);
## Evaluating the model
log_reg = log_reg.score(X_test,y_test)

print("Accuracy on the test set: {:.2f}".format(log_reg))
```

```python
## Train the model  (KNN)
knn = KNeighborsClassifier()
knn.fit(X_train,y_train);
## Evaluating the model
knn = knn.score(X_test,y_test)
print("Accuracy on the test set: {:.2f}".format(knn))
```

```python

# Train the model
clf = RandomForestClassifier()
clf.fit(X_train, y_train)

# Evaluate the model and print the score
clf_score = clf.score(X_test, y_test)
print("Accuracy on the test set: {:.2f}".format(clf_score))
```

```python
from sklearn.svm import SVC

# Train the model
svm = SVC()
svm.fit(X_train, y_train)

# Evaluate the model and print the score
svm_score = svm.score(X_test, y_test)
print("Accuracy on the test set: {:.2f}".format(svm_score))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

# Train Logistic Regression
log_reg = LogisticRegression(random_state=0)
log_reg.fit(X_train, y_train)
log_reg_score = log_reg.score(X_test, y_test)
print("Logistic Regression Accuracy: {:.2f}".format(log_reg_score))

# Train KNN
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
knn_score = knn.score(X_test, y_test)
print("KNN Accuracy: {:.2f}".format(knn_score))

# Train Random Forest
clf = RandomForestClassifier()
clf.fit(X_train, y_train)
clf_score = clf.score(X_test, y_test)
print("Random Forest Accuracy: {:.2f}".format(clf_score))

# Train SVM
svm = SVC()
svm.fit(X_train, y_train)
svm_score = svm.score(X_test, y_test)
print("SVM Accuracy: {:.2f}".format(svm_score))
```

```python
model_compare = pd.DataFrame({
    "Logistic Regression": [log_reg],
    "KNN": [knn],
    "Random Forest Classifier": [clf],
    "Support Vector Machine": [svm],
}, index=["accuracy"])
```

```python
# Create a hyperparameter grid for LogisticRegression
log_reg_grid = {"C": np.logspace(-4, 4, 20),"solver": ["liblinear"]}
# Tune LogisticRegression
np.random.seed(42)
# Setup random hyperparameter search for LogisticRegression
rs_log_reg = RandomizedSearchCV(LogisticRegression(),
                                  param_distributions=log_reg_grid,
                                  cv=5,
                                  n_iter=20,
                                  verbose=True)
# Fit random hyperparameter search model for LogisticRegression
rs_log_reg.fit(X_train, y_train)
score = rs_log_reg.score(X_test,y_test)
print(score*100)
```

```python
log_reg_grid = {'C': np.logspace(-4,4,30),
"solver":["liblinear"]}
#setup  the gird cv
gs_log_reg = GridSearchCV(LogisticRegression(),
                          param_grid=log_reg_grid,
                          cv=5,
                           verbose=True)
#fit grid search cv
gs_log_reg.fit(X_train,y_train)
score = gs_log_reg.score(X_test,y_test)
print(score*100)
```

```python
y_preds = gs_log_reg.predict(X_test)
y_preds
```

```python
sns.set(font_scale=2)
import seaborn as sns
sns.heatmap(confusion_matrix(y_test,y_preds), annot=True,cbar=False, fmt='g')
plt.xlabel("True label")
plt.ylabel("Predicted label");
```

```python
print(accuracy_score(y_test,y_preds))
```

```python
print(classification_report(y_test, y_preds))
```
